```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

### The diabetes dataset

El conjunto [diabetes dataset](https://scikit-learn.org/stable/datasets/toy_dataset.html#diabetes-dataset) incluye 442 muestras de datos sobre la diabetes, con 10 variables características, algunas de las cuales incluyen:

- Edad: Edad en años
- IMC: Índice de Masa Corporal
- PA: presión arterial promedio
- s1 tc: Células T (un tipo de glóbulos blancos)

1. Cargar los datos X e Y.

```python
X, y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])
```

2. Seleccionar una parte de este conjunto de datos seleccionando la 3ª columna del conjunto de datos. Utilizando el operador `:` para seleccionar todas las filas y, a continuación, seleccionando la 3ª columna utilizando el índice (2). También se puede cambiar la forma de los datos para que sean una matriz 2D, según sea necesario para el plotting, utilizando `reshape(n_rows, n_columns)`. Si uno de los parámetros es -1, la dimensión correspondiente se calcula automáticamente.

```python
X = X[:, 2]
X = X.reshape((-1,1))
```

3. Dividir los datos (X) y el target (Y) en conjuntos de prueba y entrenamiento. Scikit-learn tiene una forma sencilla de hacer esto

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

4. Ahora estamos listos para entrenar el modelo. Cargar el modelo de regresión lineal y entrenarlo con el conjunto de entrenamiento utilizando `model.fit()`

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

A continuación, creamos una predicción utilizando datos de prueba, utilizando la función `predict()`. Esto se usará para plotear la línea entre los grupos de datos

```python
y_pred = model.predict(X_test)
```

Ahora es el momento de mostrar los datos en un gráfico. 

`Matplotlib` es una herramienta muy útil para esta tarea. Cree un diagrama de dispersión de todos los datos de prueba X e Y, y utilice la predicción para dibujar una línea en el lugar más apropiado, entre las agrupaciones de datos del modelo.


```python
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
plt.show()
```

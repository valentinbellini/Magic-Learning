```python
import pandas as pd

df = pd.DataFrame()

df = pd.read_csv('./data/sales_data.csv')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import norm


df_uk = df[(df['Country'] == 'United Kingdom') & (df['Quantity'] > 0) & (df['Description'] != '?') & (df['UnitPrice'] > 0)]

precios = df_uk['UnitPrice']

# Calcular la media y la desviacin estndar de los precios
media = np.mean(precios)
desviacion_estandar = np.std(precios)


# Crear un rango de valores para los precios
x = np.linspace(media - 3*desviacion_estandar, media + 3*desviacion_estandar, 100)

# Calcular la funcin de densidad de probabilidad (PDF) de la distribucin normal
pdf = norm.pdf(x, media, desviacion_estandar)

# Crear el grfico
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, color='blue', linewidth=2, label='Distribucin Normal')
plt.title('Campana de Gauss de los Prsecios')
plt.xlabel('Precio')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)
plt.legend()
plt.show()
```
